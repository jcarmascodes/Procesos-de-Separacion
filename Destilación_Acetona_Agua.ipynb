**Ecuaciones de NRTL**
\begin{equation}
\ln\gamma_{1}=x_{2}^{2}[\tau_{21}(\frac{G_{21}}{x_{1}+x_{2}G_{21}})^{2}+\frac{G_{12}\tau_{12}}{(x_{2}+x_{1}G_{12})^{2}}]
\end{equation}

\begin{equation}
\ln\gamma_{2}=x_{1}^{2}[\tau_{12}(\frac{G_{12}}{x_{2}+x_{1}G_{12}})^{2}+\frac{G_{21}\tau_{21}}{(x_{1}+x_{2}G_{21})^{2}}]
\end{equation}

\begin{equation}
G_{12}=\exp(-\alpha\tau_{12})
\end{equation}

\begin{equation}
G_{21}=\exp(-\alpha\tau_{21})
\end{equation}

\begin{equation}
\tau_{12}=\frac{b_{12}}{RT}
\end{equation}

\begin{equation}
\tau_{21}=\frac{b_{21}}{RT}
\end{equation}

```python
# DATOS
R=1.9872       #cal/mol*K  BTU/lbmol*K

#p=0.001          # bar
#P=p*750
P = 1.5

#X1=np.arange(0,1.01,0.01)
X1 = np.arange(0.01,1.,0.01)
# Ecuacin de Antoine
# ACETONA
comp1 = 'ACETONA'
A1=4.42448
B1=1312.253
C1=-32.445

#A1=5.33675
#B1=1648.22
#C1=230.918

# AGUA
comp2 = 'AGUA'
A2=4.6543
B2=1435.264
C2=-64.848

#A2=5.111564
#B2=1687.537
#C2=230.170

tVAP1=(B1/(A1-math.log(P,10)))-C1
tVAP2=(B2/(A2-math.log(P,10)))-C2


tVAP1=float(tVAP1)
tVAP2=float(tVAP2)

if tVAP1<tVAP2:
    tmin = tVAP1
    tmax = tVAP2
    print('Tmin = %.4f' %tVAP1,comp1)
    print('Tmax = %.4f' %tVAP2,comp2)
if tVAP1>tVAP2:
    tmin = tVAP2
    tmax = tVAP1
    print('Tmin = %.4f' %tVAP2,comp2)
    print('Tmax = %.4f' %tVAP1,comp1)

#tmin=tmin-150
#tmax=tmax+150

#print('Tmin = ', tmin)
#print('Tmax = ', tmax)

alfa=0.3 #Se especifican los parmetros de NRTL para la mezcla de estudio
a12=-0.8009
b12=246.18
a21=3.4578
b21=-586.081

X2=np.zeros([0])
G1=np.zeros([0])
G2=np.zeros([0])
GEX_RT=np.zeros([0])
P1=np.zeros([0])
P2=np.zeros([0])
val_t=np.zeros([0])

for x1 in X1: #Se aplican las ecuaciones de NRTL para obtener los coeficientes de actividad, iterando la temperatura
    x2=1-x1

    #Ti=tmax+273.15
    Ti=tmax
    TOL=1e-5
    DT=1


    while DT>TOL:

        T=Ti
        tau12=a12+(b12/T)
        tau21=a21+(b21/T)
        G12=math.exp(-alfa*tau12)
        G21=math.exp(-alfa*tau21)

        S1=tau21*(G21/(x1+x2*G21))**2
        D1=tau12*G12/(x2+x1*G12)**2
        ln_g1=x2**2*(S1+D1)

        S2=tau12*(G12/(x2+x1*G12))**2
        D2=tau21*G21/(x1+x2*G21)**2
        ln_g2=x1**2*(S2+D2)


        g1=math.exp(ln_g1)
        g2=math.exp(ln_g2)
        gex_rt1=x1*ln_g1+x2*ln_g2

        f= lambda t: (x1*g1*10**(A1-(B1/(C1+t))))+(x2*g2*10**(A2-(B2/(C2+t))))- P
        t=op.bisect(f,100,800,xtol=TOL)

        #Ti1=t+273.15
        Ti1=t
        DT=abs(Ti-Ti1)
        Ti=Ti1


    X2=np.append(X2,x2)
    G1=np.append(G1,g1)
    G2=np.append(G2,g2)
    GEX_RT=np.append(GEX_RT,gex_rt1)

    val_t=np.append(val_t,t)
    p1=x1*g1*10**(A1-(B1/(C1+t)))
    p2=x2*g2*10**(A2-(B2/(C2+t)))

    P1=np.append(P1,p1)
    P2=np.append(P2,p2)


PTOT=P1+P2
Y1=P1/PTOT
Y2=P2/PTOT
tI=(val_t*1.8)+32
TI=tI+459.67
K1=Y1/X1
K2=Y2/X2


NRTL=pd.DataFrame()
NRTL['$x_1$']=X1
NRTL['$x_2$']=X2
NRTL['$\gamma_1$']=G1
NRTL['$\gamma_2$']=G2
NRTL['$P_1$ [bar]']=P1
NRTL['$P_2$ [bar]']=P2
NRTL['$P_{tot}$ [bar]']=PTOT
NRTL['T [K]']=val_t
#NRTL['T [R]']=TI
NRTL['$y_1$']=Y1
NRTL['$y_2$']=Y2
NRTL['$K_1$']=K1
NRTL['$K_2$']=K2
NRTL['$\alpha$']=K1/K2



ALFA=K1/K2         # Volatilidad relativa

alfa_coefx=np.polyfit(X1,ALFA,10) #Se ajustan los puntos por medio de un ajuste polinmico
alfa_coefy=np.polyfit(Y1,ALFA,10)


#alfa_polix=np.polyval(alfa_coefx,x)
#alfa_poliy=np.polyval(alfa_coefy,y)


# DATOS DE EQUILIBRIO----------------------------------------------
x=np.arange(0,1.01,0.01)
alfa_x=np.polyval(alfa_coefx,x)
y=(alfa_x*x)/(1+x*(alfa_x-1))

#------------------------------------------------------------------
plt.plot(x,y)
plt.plot(X1,Y1)
plt.show()

NRTL[50:]
```

Se desea separar una mezcla de lquido saturado compuesta por acetona-benceno que se alimenta a 72.51 kmol/h.
* *Fraccin en alimentacin*: 0.44
* *Pureza destilado*: 0.999
* *Pureza fondos: 0.001
* *Razn de reflujo*: 2.78

De acuerdo a los resultados del Mtodo de NRTL, la temperatura de burbuja de la mezcla es de 58C.


```python
comp1='Acetona'
comp2='Agua'
# Acetona , Agua
C1 = [4.215e+7, 5.2053e+7]
C2 = [0.3397, 0.3199]
C3 = [0, -0.212]
C4 = [0, 0.25795]
Tc = [562.05, 631]         # Temperatura crtica en K
pm = [58.08 , 18.01528]       # Peso molecular
T = 273.15+58              # Temperatura del sistema en K

tF=58     #C
tburbuja = 58
z=0.44 #Fraccin de alimentacin
F=72.51   # kmol/h
R=2.78 #Razn de reflujo
xD=0.999 #Pureza del destilado
xB=0.001 #Pureza de los fondos

#BALANCE DE MATERIA
B = ((F*z)-(F*xD))/(xB-xD)    # fondos
D = F-B                       # destilado

n = len(C1)
Hv = np.zeros(n)
CL = np.zeros(n)
Tr = np.zeros(n)

for i in range (n):
    Tr[i] = T/Tc[i]
    Hv[i] = C1[i]*(1-Tr[i])**(C2[i] + C3[i]*Tr[i] + C4[i]*Tr[i]**2)
    CL[i] = Hv[i]/4184

CL_prom = (z*CL[0]) + ((1-z)*CL[1])
PMF = 100 / ((z*100/pm[0])+(1-z)*100/pm[1]) #Clculo del peso molecular promedio de la mezcla
Lmda = CL_prom/PMF  # Calor latente de vaporizacin de la alimentacin (cal/g)

def cp_acetona(T): # cal/gC
  cof=[135600,-177,0.2837,0.000689]
  cp=(cof[0]+(cof[1]*T)+(cof[2]*(T**2))+(cof[3]*(T**3)))/(pm[0]*4184) #Polinomios Cp del componente 1
  return (cp)
def cp_agua(T):
  cof=[276370,-2090.1,8.125,-0.014116,9.37E-6]
  cp=(cof[0]+(cof[1]*T)+(cof[2]*(T**2))+(cof[3]*(T**3))+(cof[4]*(T**4)))/(pm[1]*4184) #Polinomios Cp del componente 2
  return (cp)

cp1 = cp_acetona(T)
cp2 = cp_agua(T)
cptotal = (cp1+cp2)/2 #Se obtiene el Cp de la mezcla al promediar el Cp de los componentes puros


q = 1 + ((cptotal*(tburbuja-tF))/Lmda) #Clculo de q para una alimentacin fra (no saturada)
print(q)


# CLCULO DE PUNTO PINCH
#lnea de referencia
if q==1:   #Lquido saturado
    xP=z
    alfa=np.polyval(alfa_coefx,xP)
    yP=(alfa*xP)/(1+xP*(alfa-1))

if q==0:   #Vapor saturado
    yP=z
    alfa=np.polyval(alfa_coefy,yP)
    xP=yP/(alfa-(yP*(alfa-1)))
#-----------------------------------
else:
    i=1
    xit=0
    tol=0.0005
    e=1
    while e>tol:
        alfa=np.polyval(alfa_coefx,xit)
        xP=((alfa-1)*(z+q)-alfa+(((alfa-1)*(z+q)-alfa)**2+(4*z*(alfa-1)*q))**0.5)/(2*(alfa-1)*q)
        yP=(alfa*xP)/(1+xP*(alfa-1))
        alfa1=np.polyval(alfa_coefx,xP)
        e=abs(alfa1-alfa)
        xit=xP
        i=i+1
#----------------------------------------

xpinch=[z,xP]
ypinch=[z,yP]
print(xpinch)
print(ypinch)

xint=(z*(R+1)+xD*(q-1))/(q*(R+1)-R*(q-1)) #Clculo del punto de interseccin
yint=(R*xint/(R+1))+(xD/(R+1)) #Clculo de la coordenada y de interseccin a partir de la lnea de operacin de rectificacin
S=(xint-xB)/(yint-xint) #Clculo de la razn S


V=S*B #V: Vapor, #S: Razn de recirculacin Reboiler, #B: Flujo de fondos
L=R*D #L: Lquido #R: Razn de reflujo condensador  #D: Flujo de destilado

print('Flujo del destilado: %f kmol/h' %D)
print('Flujo residual: %f kmol/h' %B)
print('Flujo de vapor: %f kmol/h' %V)
print('Flujo de lquido: %f kmol/h' %L)


ms=(xD-yint)/(xD-xint)
xrec=np.linspace(xint,xD,50) #Clculo de la lnea de operacin de la zona de rectificacin
yrec=(ms*xrec)+(xD/(R+1))
mi=(yint-xB)/(xint-xB)
xag=np.linspace(xB,xint,50) #Clculo de la lnea de operacin de la zona de agotamiento
yag=(mi*xag)-(xB/S)


xr=xD
yr=xD
XR=np.array([xr])
YR=np.array([yr])
i=0

while yr>yint: #Se pone de condicin el punto de interseccin donde debe frenarse este ciclo
    alfa=np.polyval(alfa_coefy,yr)
    xr=yr/(alfa-yr*(alfa-1)) #Se calcula el nmero de platos evaluando el punto de interseccin del ajuste de datos de equilibrio en "x"
    XR=np.append(XR,xr)
    YR=np.append(YR,yr)
    yr=(ms*xr)+(xD/(R+1)) #Se evala el punto de interseccin con la lnea de operacin de rectificacin para encontrar "y"
    XR=np.append(XR,xr)
    YR=np.append(YR,yr)

    i=i+1
    #xr=XR[i]
    #yr=YR[i]

p_alim=i #Se pone un contador del nmero de platos para determinar el plato de alimentacin
xa=XR[-2]
ya=YR[-2]
XA=np.array([xa])
YA=np.array([ya])




while ya>xB: #Se pone de condicin para frenar el ciclo, que el clculo se haga entre el punto de interseccin y la fraccin de componente en los fondos
    ya=(mi*xa)-(xB/S) #Se empieza calculando "y" al evaluar la lnea de operacin de agotamiento
    XA=np.append(XA,xa)
    YA=np.append(YA,ya)
    if ya<xB:
        break
    alfa=np.polyval(alfa_coefy,ya) #Posterior a ello, se evala el ajuste de datos de equilibrio para encontrar "x"
    xa=ya/(alfa-ya*(alfa-1))
    XA=np.append(XA,xa)
    YA=np.append(YA,ya)
    i=i+1 #Se contnua contando en  no. de platos hasta finalizar el ciclo

plt.figure(1,figsize=(8,6))
plt.plot(x,x,'k--')
plt.plot(x,y)
plt.plot(xpinch,ypinch)
plt.plot(xrec,yrec)
plt.plot(xag,yag)
plt.plot(XR,YR)
axhline(xD, color='gray', ls="dotted", label = 'YD')
axvline(xD, color='darkcyan', ls="dotted", label = 'XD')
axvline(xpinch[0], color='purple', ls="dotted", label = 'XF')
axvline(xB, color='c', ls="dotted", label = 'XR')
plt.plot(XA,YA)

plt.title('Determinacin de platos tericos mezcla %s - %s ' %(comp1,comp2))
plt.xlabel('x',fontsize=14)
plt.ylabel('y',fontsize=14)
plt.minorticks_on()
plt.grid( which='minor',linestyle='dotted', linewidth=1)
plt.grid(which='major')
plt.xlim(0,1)
plt.ylim(0,1)
plt.savefig('torre1.png')


print('Nmero de etapas: %i' %i)
print('Plato de alimentacin: %i' %p_alim)
```
